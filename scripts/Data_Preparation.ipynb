```python
import pandas as pd


dresses = pd.read_csv("C:/Users/chira/Desktop/Fashion Visual Search & Intelligent Styling Assistant/Dataset/dresses_bd_processed_data.csv")
jeans = pd.read_csv("C:/Users/chira/Desktop/Fashion Visual Search & Intelligent Styling Assistant/Dataset/jeans_bd_processed_data.csv")
```

```python
# Add category_type column
dresses['category_type'] = 'dress'
jeans['category_type'] = 'jeans'
```

```python
df = pd.concat([dresses, jeans], ignore_index=True)
```

```python
df
```

```python
import ast

def extract_price(price_str):
    try:
        price_dict = ast.literal_eval(price_str)
        return list(price_dict.values())[0]
    except:
        return None  # or 0

df['selling_price_clean'] = df['selling_price'].apply(extract_price)
df['mrp_clean'] = df['mrp'].apply(extract_price)
```

```python
df
```

```python
def clean_text(text):
    if pd.isnull(text):
        return ""
    return str(text).replace('\n', ' ').replace('[', '').replace(']', '').replace('{', '').replace('}', '').strip()

text_columns = ['meta_info', 'feature_list', 'description', 'style_attributes']

for col in text_columns:
    if col in df.columns:
        df[col + '_clean'] = df[col].apply(clean_text)
```
